### K-Means using SK-learn 



```python
from sklearn.cluster import KMeans
```

```python
KModule = KMeans()
```

```python
K_model = KModule.fit(Df)
```

```python
K_result = K_model.predict(Df)
```

```python
centers = K_model.cluster_centers_
```
