## building ML algorithm for ‘tenure’ change/vary w.r.t change in ‘TotalCharges’ of the customer.

```python
# divide dependent and independent parameters 
x = pd.DataFrame(Df['MonthlyCharges'])
y = pd.DataFrame(Df['tenure'])
```

```python
#split the Dataset by 70%

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size= 0.3, random_state =123)
```

```python
#prepare model 
from sklearn.linear_model import LinearRegression
lm= LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
y_result = lm.predict(x_test)
```

```python
plt.scatter( y_result,y_test)
plt.xlabel("predicted results")
plt.ylabel("Test data")
```

```python
Error = y_result - y_test
```

```python
sns.distplot(Error)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error (y_test,y_result)
```

```python
metrics.mean_squared_error (y_test,y_result)
```

```python
np.sqrt(metrics.mean_squared_error (y_test,y_result))
```
